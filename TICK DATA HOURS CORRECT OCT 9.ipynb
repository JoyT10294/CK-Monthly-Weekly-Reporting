```python
#Import useful functions
import warnings
warnings.filterwarnings("ignore")
import itertools
import xlrd
import xlwt
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import pylab as plt
import numpy as np
import xlsxwriter
from xlrd import open_workbook
from datetime import datetime
plt.style.use('fivethirtyeight')
```

```python
#Designate filename of original tick data excel file
filenameorig = '/Users/JoyThompson/Desktop/Tick Data/tickdatadec.xlsx'
#Read the excel file and write to a pandas dataframe
xl2pd = pd.read_excel(filenameorig, sheetname='tickdata')
#Set the index of the tick data dataframe to the index to parse through all data values
xl2pd=xl2pd.set_index('Id')
#Designate the filename of the excel file to write the new dataframe to
filenamefinal= '/Users/JoyThompson/Desktop/Tick Data/JoyTickData.xlsx'
#Initialize excel writer to open and write to the above designated file
writer = pd.ExcelWriter(filenamefinal, engine='xlsxwriter')
```

```python
#Create a new dataframe from the orignal tick dataframe using the date column as the index
xl2pddate=xl2pd.set_index('Date')
#Sort the dataframe by the indexed date column
xl2pddate=xl2pddate.sort_index(axis=0)
#Create a new dataframe by grouping the above dataframe data by month
dategroups=xl2pddate.groupby(pd.Grouper(freq='M'),).nunique()  
#Initiate a list of strings which will serve as the first column headers
dfcol=['Id','Project','Project Total','Task','Task Total','User', 'User Total']
#Create a list of the months as strings
monthstringvec=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
#Loop through the index of the dataframe sorted by month, transform the data to appropriate string, and append it
#to the list of column headers initiated above
for datamonths in dategroups.index:
    monthdate=monthstringvec[datamonths.month-1]+' ' +str(datamonths.year)
    dfcol.append(monthdate)
```

```python
comblist=list()
for ind in xl2pd.index:
    project=xl2pd.loc[ind,'Project']
    task=xl2pd.loc[ind,'Task']
    user=xl2pd.loc[ind,'User']
    comb=[project, task, user]
    if comb not in comblist:
        comblist.append(comb)
```

```python
df=pd.DataFrame({},columns=dfcol)
df=df.set_index('Id')
ind=0
for comb in comblist:
    df.loc[ind,'Project']=comb[0]
    df.loc[ind,'Task']=comb[1]
    df.loc[ind,'User']=comb[2]
    ind=ind+1
```
